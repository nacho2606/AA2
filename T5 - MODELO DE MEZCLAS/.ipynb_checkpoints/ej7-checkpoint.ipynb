```python
import numpy as np
from sklearn.decomposition import LatentDirichletAllocation
import matplotlib.pyplot as plt
from time import time
```

## Latent Dirichlet Allocation (sección 20.6.1)

Se trata de un modelo generativo que permite que cada muestra pueda pertenecer a más de un clúster. Siguiendo la fig. 20.14, el modelo tiene esta forma:
$$
\alpha \rightarrow \pi^n \rightarrow z^n_w \rightarrow v^n_w \leftarrow \theta
$$
Para interpretarlo, vamos a suponer que trabajamos con $N$ documentos y $W_n$ es el número de palabras del documento $n$. En este contexto, $v^n_w$ representa la palabra del documento $n$ que aparece en la posición $w$.

Un posible proceso generativo de las palabras de todos los documentos podría ser:
- Se elige un tema $z^n_w$ de entre la lista de temas posibles. Puesto que un documento podría tratar sobre varios asuntos, puede contener palabras que pertenezcan a distintos temas. La elección de $z^n_w$ se hace utilizando $\pi^n$, que asigna distintas probabilidades a cada tema.
- Una vez se tiene el tema, se escoge una palabra de entre todas las disponibles para este tema por medio de la distribución $\theta$. En concreto, la probabilidad de seleccionar la palabra $i$ dado el tema $k$ vendría dada por $\theta_{i|k}$.
  
Además, la variable $\alpha$ modela la distribución de temas. Para cada documento, los temas concretos que pueden aparecer en él vienen dados por $\pi^n$ que ha sigo elegida a partir de la distribución $\alpha$.

## Ejemplo 15.6.

Un ejemplo que ya hemos visto que se presta a esta modelización es el 15.6, que se resolvió en la tarea 3 utilizando PLSA. Dado un tema $k$, por ejemplo el relativo a las mascotas, la distribución de palabras que pueden usarse viene dada por el correspondiente array pi; en el caso de las mascotas, por p4. Puesto que este es un ejemplo artificial nosotros conocemos las distribuciones p1, p2, p3 y p4, las utilizamos para generar el conjunto de datos y podemos así analizar los resultados obtenidos. Sin embargo, el algoritmo LDA no las conoce y las tiene que aprender, al igual que $\alpha$ (que está relacionado con el número de temas). Se trata aquí de replicar el ejercicio de la tarea 3 aplicando LDA. Entiéndase lo que ocurre y el significado de las distintas variables y piénsese en las diferencias entre ambos enfoques.

[En este ejemplo cada palabra aparece una vez o ninguna en el documento, pero el algoritmo funciona igual con un número arbitrario de ocurrencias).]

En primer lugar, recuperamos parte del código que se utilizó en la tarea 3. Primero, la generación del conjunto de datos.

```python
p1 = np.array([0, 1, 0.8, 0.8, 0.7, 0.5]) # [influenza, flu, headache,nose,temperature,bed] probabilidad de ocurrir
p2 = np.array([1, 0, 0.8, 0.8, 0.7, 0.5]) # [influenza, flu, headache,nose,temperature,bed] probabilidad de ocurrir

p1 = np.concatenate((p1, 0.05*np.ones(4))) # añadir palabras aleatorias con pocas ocurrencias
p2 = np.concatenate((p2, 0.05*np.ones(4))) # añadir palabras aleatorias con pocas ocurrencias
p3 = 0.1*np.ones(p1.size) # tema de fondo aleatorio con pocas ocurrencias
p4 = np.array([0, 0.1, 0.1, 0.1, 0, 0.2, 0.8, 0.8, 0.9, 0.9]) # otro tema sobre 'mascotas'

D = p1.size # número de palabras en el diccionario
N = 500 # número de documentos a muestrear de cada tipo
rng1 = np.random.default_rng()
x1 = (rng1.random((D,N)).T < p1).T
x2 = (rng1.random((D,N)).T < p2).T
x3 = (rng1.random((D,N)).T < p3).T
x4 = (rng1.random((D,N)).T < p4).T

X = np.concatenate((x1, x2, x3, x4), axis=1) # incluye el tema de fondo
# las columnas de X corresponden a datos de entrada
X.shape
```

```python
plt.figure(figsize = (13,2))
plt.imshow(X, aspect='auto', cmap='gray_r', interpolation='none')
```

Esta rutina nos permitía dibujar diagramas de Hinton.

```python
def hinton(matrix, max_weight=None, ax=None):
    """Draw Hinton diagram for visualizing a weight matrix."""
    ax = ax if ax is not None else plt.gca()

    if not max_weight:
        max_weight = 2 ** np.ceil(np.log2(np.abs(matrix).max()))

    ax.patch.set_facecolor('gray')
    ax.set_aspect('equal', 'box')
    ax.xaxis.set_major_locator(plt.NullLocator())
    ax.yaxis.set_major_locator(plt.NullLocator())

    for (x, y), w in np.ndenumerate(matrix):
        color = 'white' if w > 0 else 'black'
        size = np.sqrt(np.abs(w) / max_weight)
        rect = plt.Rectangle([x - size / 2, y - size / 2], size, size,
                             facecolor=color, edgecolor=color)
        ax.add_patch(rect)

    ax.autoscale_view()
    ax.invert_yaxis()
```

En este punto es donde aplicamos LDA. Primero, con solo dos componentes/temas.

```python
n_topics = 2
lda = LatentDirichletAllocation(n_components = n_topics)
# Entrenamos y nos quedamos con la proporción de cada tema en cada documento.
patterns = lda.fit_transform(X.T)
# Obtenemos los temas.
topics = lda.components_
```

```python
hinton(topics)
```

```python
plt.figure(figsize = (13,2))
plt.imshow(patterns.T, aspect='auto', cmap='gray')
```

Ahora, con 3 temas.

```python
n_topics = 3
lda = LatentDirichletAllocation(n_components = n_topics)
# Entrenamos y nos quedamos con la proporción de cada tema en cada documento.
patterns = lda.fit_transform(X.T)
# Obtenemos los temas.
topics = lda.components_
```

```python
hinton(topics)
```

```python
plt.figure(figsize = (13,2))
plt.imshow(patterns.T, aspect='auto', cmap='gray')
```

###Comparativa resultados vs Ejercicio 2 Tarea 4

#### Objetivo y Modelo:

- LDA (Ejercicio Actual): Busca descubrir temas latentes en los documentos, usando distribuciones de Dirichlet para obtener temas coherentes y bien definidos.

- PLSA y PCA (Ejercicio 2): PCA se utiliza para reducir la dimensionalidad y simplificar los datos. PLSA, similar a LDA, modela temas latentes, pero sin la regularización de Dirichlet, lo que puede llevar a temas más específicos pero menos generalizables.

####Resultados de Análisis Temático:

- LDA: Genera temas claros con distribuciones de palabras que tienden a ser interpretables debido a la regularización.

- PLSA: Ofrece una descomposición probabilística más ajustada a la muestra, captando relaciones sutiles, pero puede generar temas menos robustos en datos muy variados.

####Respresentaciones gráficas

En cuanto a los gráficos, se puede observar la diferencia entre modelos. La degradación y separación observada en los gráficos de PCA versus LDA y PLSA se debe a la naturaleza y objetivo de cada técnica:

- PCA: Al reducir la dimensionalidad, PCA no modela relaciones probabilísticas de pertenencia entre temas o documentos, sino que busca ejes ortogonales de máxima varianza. Así, los puntos o líneas representan de manera clara y directa la estructura del dataset, sin probabilidades graduales, resultando en líneas o áreas bien delimitadas y sin suavización.

- LDA y PLSA: Debido a esta probabilidad compartida, los gráficos de LDA y PLSA muestran gradientes o transiciones suaves entre temas, ya que cada documento no se asigna exclusivamente a un único tema, sino que se representa en función de una distribución que degrada la delimitación clara entre temas.
