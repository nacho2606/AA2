## Mezclas de expertos.

Barber los menciona pero no los desarrolla. Se pueden encontrar en la sección 14.5 de Bishop, 12.9 de Theodoridis (replica Bishop) y 13.6.2 del primer volumen de Murphy.

Aquí implementamos el ejemplo ilustrado en la figura 14.8 de Bishop pero, en lugar de utilizar el algoritmo EM, nos aprovecharemos de las facilidades que ofrece NumPyro. Primero definimos el modelo. En lugar de utilizar variables latentes explícitas vamos a utililizar la distribución MixtureSameFamily que construye un modelos de mezclas a partir de los coeficientes de la mezcla y las densidades de las componentes.

```python
def mixture_of_linear_regressions(X, y=None):
    # Parámetros para la primera regresión lineal.
    theta1 = sample('theta1', dist.Normal(0, 2))
    b1 = sample('b1', dist.Normal(0, 3))

    # Parámetros para la segunda regresión lineal.
    theta2 = sample('theta2', dist.Normal(0, 2))
    b2 = sample('b2', dist.Normal(0, 3))

    # Ruido compartido
    #sigma = sample('sigma', dist.Exponential(1.0))
    sigma = np.sqrt(0.1)  # Ruido real metido directamente.
                          # Sorprendentemente, cuando se utiliza el valor real en vez de dejar que
                          # NumPyro lo aprenda es habitual que los resultados no sean buenos.

    # Medias (valores esperados) para los dos modelos.
    mean1 = theta1 * X + b1
    mean2 = theta2 * X + b2

    # Apilamos las dos medias para poder utilizar MixturesSameFamily.
    means = jnp.stack([mean1, mean2], axis=-1)

    # Pesos de la mezcla (probabilidad de elegir el modelo 1).
    pi = numpyro.sample("pi", dist.Uniform(0.0, 1.0))
    mix = dist.MixtureSameFamily(
        dist.Categorical(probs=jnp.array([pi, 1-pi])),
        dist.Normal(means, sigma)
    )

    # Ahora muestreamos.
    with plate('data', X.shape[0]):
        # EL segundo argumento de MixtureSameFamily tiene que ser una distribución, no una lista
        # de distribuciones. Utilizaremos una normal con la media calculada antes.
        y_obs = numpyro.sample('y_obs', mix, obs=y)

    return y_obs
```

Generamos datos sobre los que ejecutar nuestro algoritmo de aprendizaje y los mostramos en un gráfico, similar a la figura 14.8 de Bishop pero con más puntos en los segmentos uno y tres para ver cómo afecta a las probabilidades asignadas a los expertos.

```python
# Simulamos datos a partir de los modelos reales.
theta1_t = 0.2  # pendiente
b1_t = 1
theta2_t = -0.02  # pendiente
b2_t = -1
sigma_true = 0.1

rng = np.random.default_rng(seed=6)
X1 = np.linspace(-1,-0.5,15)
X2 = np.linspace(-0.5,0.5,10)
X3 = np.linspace(0.5,1,15)

y1 = X1 * theta2_t + b2_t + rng.normal(scale=np.sqrt(sigma_true), size=len(X1))
y2 = X2 * theta1_t + b1_t + rng.normal(scale=np.sqrt(sigma_true), size=len(X2))
y3 = X3 * theta2_t + b2_t + rng.normal(scale=np.sqrt(sigma_true), size=len(X3))

X = np.concatenate((X1,X2,X3))
y = np.concatenate((y1,y2,y3))
sortedX = np.argsort(X)
plt.scatter(X, y)
plt.show()
```

## MCMC.
Inferencia por medio de MCMC.

```python
nuts_kernel = NUTS(mixture_of_linear_regressions)
mcmc = MCMC(nuts_kernel, num_samples=1000, num_warmup=500)
rng_key = random.PRNGKey(0)
mcmc.run(rng_key, X, y)
samples = mcmc.get_samples()
```

Calculamos los parámetros como las medias de los valores muestreados.

```python
t1 = samples["theta1"].mean()
t2 = samples["theta2"].mean()
b1 = samples["b1"].mean()
b2 = samples["b2"].mean()
```

Dibujamos los resultados.

```python
gridX = np.linspace(-1,1,50)
out1 = t1* gridX + b1
out2 = t2* gridX + b2
plt.plot(gridX, out1)
plt.plot(gridX, out2)
plt.scatter(X, y)
```

Obtenemos las probabilidades asociadas a los expertos y las utilizamos para dibujar función de densidad de la probabilidad.

```python
pi = samples["pi"].mean()
```

```python
gridX = np.linspace(-1, 1, 50)
gridY = np.linspace(-1.5, 1.5, 50)
d1 = dist.Normal(t1 * gridX + b1, np.sqrt(sigma_true)) # densidad asociada al primer experto
d2 = dist.Normal(t2 * gridX + b2, np.sqrt(sigma_true)) # densidad asociada al segundo experto
```

```python
# Cada columna en la imagen que construimos se corresponde con la función de densidad
# correspondiente al punto x: y ~ pi1 N(y | f1(x)) + pi2 N(y | f2(x))
image = np.zeros((50, 50))
for i in range(50):
    image[49-i,:] = pi * np.exp(d1.log_prob(gridY[i])) + (1-pi) * np.exp(d2.log_prob(gridY[i]))
```

Puesto que hemos incluido muchos más puntos en la recta inferior, la densidad es mucho mayor en esta zona. Si el número de puntos en ambas rectas fuera el mismo, la gráfica sería como la del libro de Bishop.

```python
plt.imshow(image)
```

## Dependencias respecto de la entrada.
En el modelo anterior, las probabilidades asociadas a los expertos no dependen del dato de entrada x. Así, dado un nuevo x no está claro si deberíamos utilizar el prmer experto o el segundo para predecir el valor de y. Vamos a modificar el modelo para que las probabilidades dependan de la entrada siguiendo a Murphy (13.6.2.1). Ahora, la probabilidad de escoger el experto $z = k$ viene dada por
$$
p(z = k) = \textrm{Cat}(z \mid \textrm{softmax}(\mathbf{Vx}))
$$
En nuestro ejemplo vamos a implementar una función que sea cuadrática en x: puesto que trabajamos con 2 expertos, la dimensión de la matriz será 2x3.

```python
import jax.nn
```

```python
def mixture_of_linear_regressions2(X, y=None):
    # Construimos primero la matriz de pesos de selección V.
    weights = sample('weights', dist.Normal(0, 1).expand([2,3]))  # Para 2 expertos, 3 coeficientes.
    # Logits lineales que dependen de X.
    logits = jnp.dot(weights, np.vstack((X**2, X, np.ones(X.shape[0]))))

    # Probabilidades de selección utilizando la función softmax.
    mixing_probs = jax.nn.softmax(logits, axis=0).T

    # Parámetros para la primera regresión lineal.
    theta1 = sample('theta1', dist.Normal(0, 2))
    b1 = sample('b1', dist.Normal(0, 3))

    # Parámetros para la segunda regresión lineal.
    theta2 = sample('theta2', dist.Normal(0, 2))
    b2 = sample('b2', dist.Normal(0, 3))

    # Ruido compartido
    #sigma = sample('sigma', dist.Exponential(1.0))
    sigma = np.sqrt(0.1)  # Ruido real metido directamente.
                          # Sorprendentemente, cuando se utiliza el valor real en vez de dejar que
                          # NumPyro lo aprenda es habitual que los resultados no sean buenos.

    # Medias (valores esperados) para los dos modelos.
    mean1 = theta1 * X + b1
    mean2 = theta2 * X + b2

    # Apilamos las dos medias para poder utilizar MixturesSameFamily.
    means = jnp.stack([mean1, mean2], axis=-1)

    # Distribución de la mezcla.
    mixture_dist = dist.MixtureSameFamily(
        dist.Categorical(probs=mixing_probs),
        dist.Normal(means, sigma)
    )

    # Ahora muestreamos.
    with plate('data', X.shape[0]):
        y_obs = numpyro.sample('y_obs', mixture_dist, obs=y)

    return y_obs
```

```python
nuts_kernel = NUTS(mixture_of_linear_regressions2)
mcmc2 = MCMC(nuts_kernel, num_samples=1000, num_warmup=500)
rng_key = random.PRNGKey(0)
mcmc2.run(rng_key, X, y)
samples = mcmc2.get_samples()
```

Recuperamos los pesos de la matriz de selección V y los utilizamos para ver la probabilidad de elegir un experto u otro en función del dato de entrada.

```python
w = samples["weights"].mean(axis=0)
w.shape
```

```python
mixing_p = jax.nn.softmax(jnp.dot(w, np.vstack((X**2, X, np.ones(X.shape[0])))), axis=0)
mixing_p.shape
```

Como se puede apreciar, la probabilidad de asignar el primer experto a los puntos en los segmentos [-1,-0.5] y [0.5,1] es mucho menor que para los puntos en [-0.5,0.5].

```python
plt.scatter(X, mixing_p[0,:])
```

Si comparamos las rectas de regresión obtenidas, prácticamente no hay diferencias (aunque se han intercambiado las posiciones de los expertos).

```python
t1 = samples["theta1"].mean()
t2 = samples["theta2"].mean()
b1 = samples["b1"].mean()
b2 = samples["b2"].mean()

gridX = np.linspace(-1,1,50)
out1 = t1* gridX + b1
out2 = t2* gridX + b2
plt.plot(gridX, out1)
plt.plot(gridX, out2)
plt.scatter(X, y)
```

Sin embargo, la función de densidad del modelo de mezclas sí varía y cuanto más nos desplazamos hacia la izquierda menor es la importancia que se asigna al segundo experto.

```python
gridX = np.linspace(-1, 1, 50)
gridY = np.linspace(-1.5, 1.5, 50)
d1 = dist.Normal(t1 * gridX + b1, np.sqrt(sigma_true)) # densidad asociada al primer experto
d2 = dist.Normal(t2 * gridX + b2, np.sqrt(sigma_true)) # densidad asociada al segundo experto

# Cada columna en la imagen que construimos se corresponde con la función de densidad
# correspondiente al punto x: y ~ pi1(x) * N(y | f1(x)) + pi2(x) * N(y | f2(x))
image = np.zeros((50, 50))
for i in range(50):
    xin = np.array([gridX[i]**2, gridX[i], 1])
    q = jax.nn.softmax(jnp.dot(w, xin), axis=0)[0]
    image[49-i,:] = q * np.exp(d1.log_prob(gridY[i])) + (1-q) * np.exp(d2.log_prob(gridY[i]))
```

```python
plt.imshow(image)
```

### Predicción.
A la hora de predecir con este modelo, en lugar de tomar una media ponderada resulta más razonable quedarse con el experto al que nuestro modelo le asocia mayor probabilidad.

```python
import numpy as np
import jax
import matplotlib.pyplot as plt

# Supongamos que gridX ya está definido, y out1, out2 calculados.
gridX = np.linspace(-1, 1, 50)
out1 = t1 * gridX + b1
out2 = t2 * gridX + b2

# Recalcular logits para gridX
gridX_vstack = np.vstack((gridX**2, gridX, np.ones(gridX.shape[0])))
logits = np.dot(samples["weights"].mean(axis=0), gridX_vstack)  # Media de pesos muestreados

# Probabilidades de mezcla (softmax)
mixing_probs_grid = jax.nn.softmax(logits, axis=0).T

# Probabilidad del primer experto
p = mixing_probs_grid[:, 0]

# Predicción del experto más probable (basado en p)
out = np.where(p > 0.5, out1, out2)

# Visualización
plt.figure(figsize=(10, 6))
plt.scatter(gridX, out, label='Experto más probable', color='blue', marker='o')
plt.plot(gridX, p * out1 + (1 - p) * out2, label='Media ponderada', color='red', linewidth=2)
plt.legend()
plt.xlabel("X")
plt.ylabel("Predicción")
plt.title("Mixture of Linear Regressions")
plt.show()
```

```python
gridX = np.linspace(-1,1,50)
out1 = t1* gridX + b1
out2 = t2* gridX + b2

logits = np.dot(samples["weights"].mean(axis=0), gridX_vstack)  # Media de pesos muestreados

# Probabilidades de mezcla (softmax)
mixing_probs_grid = jax.nn.softmax(logits, axis=0).T

# Probabilidad del primer experto
p = mixing_probs_grid[:, 0]

# Predicción del experto más probable (basado en p)
out = np.where(p > 0.5, out1, out2)

plt.scatter(gridX, out)
plt.scatter(gridX, p*out1+(1-p)*out2)
plt.legend(['experto más probable', 'media ponderada'])
```
