## Ejercicio 11.1

```python
bn = gum.BayesNet('Printer')

fuse = bn.add(gum.LabelizedVariable('Fuse', '', 2))
drum = bn.add(gum.LabelizedVariable('Drum', '', 2))
toner = bn.add(gum.LabelizedVariable('Toner', '', 2))
paper = bn.add(gum.LabelizedVariable('Paper', '', 2))
roller = bn.add(gum.LabelizedVariable('Roller', '', 2))
burning = bn.add(gum.LabelizedVariable('Burning', '', 2))
quality = bn.add(gum.LabelizedVariable('Quality', '', 2))
wrinkled = bn.add(gum.LabelizedVariable('Wrinkled', '', 2))
pages = bn.add(gum.LabelizedVariable('Pages', '', 2))
jam = bn.add(gum.LabelizedVariable('Jam', '', 2))

# Agregar arcos entre los nodos para reflejar las dependencias entre las variables
bn.addArc(fuse, burning)
bn.addArc(fuse, wrinkled)
bn.addArc(fuse, jam)
bn.addArc(drum, quality)
bn.addArc(toner, quality)
bn.addArc(paper, pages)
bn.addArc(paper, wrinkled)
bn.addArc(paper, quality)
bn.addArc(roller, jam)
bn.addArc(roller, pages)
```

```python
# Crear red bayesiana
learner = gum.BNLearner(datos, bn)

# Usar el algoritmo EM para aprender parámetros de los datos con suavizado
learner.useSmoothingPrior(1e-3)

# Aplicar los parámetros aprendidos a la red bayesiana
learner.useEM(1e-3)

# Enseñar a la red
bn_learned = learner.learnParameters(bn)
```

```python
# Crear un motor de inferencia
inference = gum.LazyPropagation(bn_learned)

evidence = {"Wrinkled" : 0, "Burning" : 0, "Quality" : 1}

# Establecer la evidencia para el motor de inferencia
inference.setEvidence(evidence)

# Realizar la inferencia
inference.makeInference()

# Obtener la probabilidad de problema con la unidad de tambor
prob_drum_problem = inference.posterior(bn_learned.idFromName("Drum"))

# Mostrar la probabilidad del problema con la unidad de tambor
print(f"\nProbabilidad de problema con la unidad de tambor: {prob_drum_problem[1]}")
```

#Ejercicio 11.9

Para demostrar que ambas tablas tienen la misma puntuación de verosimilitud marginal, debemos analizar las probabilidades marginales de x1, ya que x2 no es observado.

La probabilidad marginal de x1 se obtiene sumando las probabilidades sobre todas las posibles realizaciones de x2. Es decir:

p(x1 = i) = \sum_{j=1}^{2} \theta_{i,j}

#### Probabilidades marginales de la primera tabla

Dada la tabla:

\begin{pmatrix}
0.3 & 0.3 \\
0.2 & 0.2
\end{pmatrix}

Las probabilidades marginales de x1 son:

p(x1 = 1) = \theta_{1,1} + \theta_{1,2} = 0.3 + 0.3 = 0.6

p(x1 = 2) = \theta_{2,1} + \theta_{2,2} = 0.2 + 0.2 = 0.4

#### Probabilidades marginales de la segunda tabla

Para la tabla:

\begin{pmatrix}
0.2 & 0.4 \\
0.4 & 0
\end{pmatrix}


Las probabilidades marginales x1 son:

p(x1 = 1) = \theta_{1,1} + \theta_{1,2} = 0.2 + 0.4 = 0.6

p(x1 = 2) = \theta_{2,1} + \theta_{2,2} = 0.4 + 0 = 0.4

### Conclusión

En ambos casos, las probabilidades marginales x1 son las mismas:


p(x1 = 1) = 0.6 , p(x1 = 2) = 0.4


Por lo tanto, ambas tablas de probabilidad tienen la misma verosimilitud marginal. Esto demuestra que la segunda tabla \begin{pmatrix} 0.2 & 0.4 \\ 0.4 & 0 \end{pmatrix} tiene la misma puntuación de verosimilitud marginal que la primera tabla dada.
